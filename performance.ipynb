## Conclusions
It is quite clear that the `FastDiscreteFrechetMatrix` algorithm is the fastest of the lot, independently of the polyline size. It is also evident that the `LinearDiscreteFrechet` algorithm is quite capable for a range of polyline sizes. As for the `FastDiscreteFrechetSparse` algorithm, it seems obvious that the "sparse matrix" implementation based on a `numba.typed.Dict` object is the performance bottleneck: it does not seem to scale well. Future versions of this class should use a more performant support for sparse matrices.
